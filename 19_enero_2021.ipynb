# slicing (continuacion)
[ inicio : stop : incremento]

notas:
1. el stop no es incluido
1. el incremento se puede modificar
1. soporta indexado alternativo (negativo)


```
frase = "como hace a la gente mas positiva? muy facil has que pierdan un electron"
print(frase)
# imprimir "ace" de la palabra hacer
print(frase [6 : 9 : 1])

#valores por defecto
print(frase[: :])

#incremento
print(frase[ : : 2]) # toda la frase
print(frase [-1 : :-1]) # inverso

nombre = "juan jose"
print(nombre[1::-1])
primer = nombre[:4:]
print(primer)
print(nombre[-9: :1])
```

# metodos de la clase string(str)
en python todo es un objeto (POO)
eso signinifica que cada tipo de dato trae consigo metodos (funciones) incluidas
para identificar los metodos disponibles se puede usar una de dos funciones:
1 dir(str) # resumen de los metodos de esa clase
2. help (str) # informacion completa de los metodos en esa clase

```
nombre = "juan jose"
print(nombre)
print(dir(str))
```

los nombre de metodos que inician con doble guion bajo significa que son metodos privados(POO), y no tendrias porque usarlos como programador consumidor.
Los demas metodos son publicos y son de libre uso

```
print(nombre)
print(nombre.islower())
nombre = "pedro fernandez"
print(nombre.capitalize())
print(nombre.endswith("dez"))
print(nombre.replace("e", "*"))
ind_e = nombre.find("e")
print(ind_e)
nombre = nombre [:ind_e: ] + "@" + nombre[ind_e+1::]
print(nombre)
print(nombre.replace("e", "@", 1))
```

```
print(help(str))
```

```
# metodos para procesar strings

numeros = "  2  ,   1, 3 , 4,   5,    45   ,7  "
print(numeros)
# como extraigo los numero de la cadena
# 1. eliminar los espacios al principio y al final con strip()
numeros = numeros.strip()
print(numeros)

#2 separo de comas
numeros = numeros.split(",")
print(numeros)
#3 con un for e indices limpiar cadenas
for indice in range(len(numeros)):
  numeros[indice] = numeros[indice].strip()
print(numeros)

#4 convertirlos a tipo entero
for indice in range(len(numeros)):
  numeros[indice] = int(numeros[indice])
print(numeros)
#5 usar los numeros



print(help(str.split))
```
